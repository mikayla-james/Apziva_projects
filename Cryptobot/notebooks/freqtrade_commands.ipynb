### Basic Fretqtrade   
This section is to get a handle of the basic freqtrade commands before creating my own strategy

````python
### Backtesting no strategy modification
!freqtrade backtesting --strategy SampleStrategy -i 1d --timerange 20250201-20250701 --export trades --export-filename 01_backtest_results.json
````

````python
### Plot Data
!freqtrade plot-dataframe --strategy SampleStrategy -p BTC/USDT -i 1d --no-trades
````

````python
### Plot Profit
!freqtrade plot-profit -i 1d
````

````python
## Backtesting with strategy modification
!freqtrade backtesting --strategy SampleStrategy -i 30m --timerange 20251001-20251201 --export trades --export-filename 02_backtest_results.json
````

````python
!freqtrade plot-dataframe -p BTC/USDT --strategy SampleStrategy -i 30m --timerange 20251001-20251201 
````
